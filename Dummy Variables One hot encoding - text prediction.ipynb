```python
import pandas as pd
df = pd.read_csv("FIFA19.csv")
df
```

```python
dummies = pd.get_dummies(df.Nationality)
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['Ukraine','Wales'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['Age','Name','Nationality','Club','Value','Wage','Preferred Foot'], axis='columns')
```

```python
Y = final.Potential
```

```python
Y
```

```python
model.fit(X,Y)
```

```python
X
```

```python
model.predict([28, 89,50])
```

```python
model.score(X,Y)
```
